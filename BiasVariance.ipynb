```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use("fivethirtyeight")
```

```python
xs = np.arange(0.0, 2.0, 0.001)
ys = np.sin(np.pi * xs)
plt.plot(xs, ys);
plt.ylim(-1.05, 1.05)
```

Use a horizontal line to fit sinusoidal function. Choose 2 random points from plot and fit based upon this, repeat for 100 trials.

```python
trials = 100
b_s = []
plt.plot(xs, ys);
plt.ylim(-1.05, 1.05)
for t in range(trials):
    indices = np.random.randint(low = 0, high = len(xs), size=2)
    rand_y = ys[indices]
    b = np.mean(rand_y)
    b_s.append(b)
    plt.axhline(y = b, alpha=0.12)
```

```python
b_avg = np.mean(b_s)
bias = np.mean((ys - b_avg)**2)
variance = np.std((ys - b_avg)**2)
plt.plot(xs, ys);
plt.ylim(-1.05, 1.05)
plt.title("Bias: %0.4f, Variance: %0.4f" %(bias, variance))
plt.fill_between(xs, b_avg - np.percentile(b_s, 0.90), b_avg + np.percentile(b_s, 0.90) ,facecolor="gray", alpha = 0.3);
plt.axhline(y = b_avg);
```

```python
from scipy.stats import linregress
from scipy import stats
```

```python
trials = 100
b_s = []
plt.plot(xs, ys);
plt.ylim(-1.05, 1.05)
slopes = []
intercepts = []
for t in range(trials):
    indices = np.random.randint(low = 0, high = len(xs), size=2)
    rand_x = xs[indices]
    rand_y = ys[indices]
    slope, intercept, r_val, pval, std_err = linregress(rand_x, rand_y)
    slopes.append(slope)
    intercepts.append(intercept)
    fitted_y = xs * slope + intercept
    plt.plot(xs, fitted_y, alpha = 0.12);
```
